```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
confusion_matrix = np.array([[79, 1,4,10,6],
                             [0,48,1,1,0],
                             [5,6,25,14,0],
                             [0,0,0,50,0],
                             [9,8,9,19,5]])
print(confusion_matrix)
```

```python
#Un-normalized confusion matrix
fig, ax = plt.subplots(figsize=(6,4))
sns.heatmap(confusion_matrix, annot=True, linewidths = 0.01, ax = ax)
ax.set_title('Unnormalized confusion matrix')
ax.set_xlabel('PREDICTED VALUES')
ax.set_ylabel('ACTUAL VALUES')

plt.show()
```

```python
#Normalized confusion matrix
#Divide each row element by the sum of the entire row
cm_normalized = confusion_matrix.astype('float') / confusion_matrix.sum(axis=1)[:, np.newaxis]

fig, ax = plt.subplots(figsize=(6,4))
sns.heatmap(cm_normalized, annot=True, linewidths = 0.01, ax = ax)
ax.set_title('Normalized confusion matrix')
ax.set_xlabel('PREDICTED VALUES')
ax.set_ylabel('ACTUAL VALUES')

plt.show()
```

```python
#Both matrices side by side to see the difference

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,6))
sns.heatmap(confusion_matrix, annot=True, ax = ax1)
ax1.set_title('Unnormalized confusion matrix')
ax1.set_xlabel('PREDICTED VALUES')
ax1.set_ylabel('ACTUAL VALUES')

sns.heatmap(cm_normalized, annot=True, ax = ax2)
ax2.set_title('Normalized confusion matrix')
ax2.set_xlabel('PREDICTED VALUES')
ax2.set_ylabel('ACTUAL VALUES')
plt.show()
```
